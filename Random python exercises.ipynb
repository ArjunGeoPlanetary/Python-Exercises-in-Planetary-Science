```python
#Define a function that takes three numbers as input parameters and returns the maximum of three numbers by using if-else statements and not using the max() function or any similar one.
# include a check whether the input values are indeed numbers. If incorrect output a message that tells the user what went wrong
# #print the result using a formatted print command

def numbers(num1,num2,num3):

  if not ((type(num1) is int or type(num1) is float) and \
          (type(num2) is int or type(num2) is float) and \
          (type(num3) is int or type(num3) is float)):
    return "Error: All inputs must be numbers"


  if num1 >= num2 and num1 >= num3:
    maximum = num1
  elif num2 >= num1 and num2 >= num3:
    maximum = num2
  else:
    maximum = num3


  print(f"The maximum of {num1}, {num2}, and {num3} is {maximum}")
  return maximum

numbers(1,2,3)
numbers(10, 5, 8)
numbers(-1, -5, 0)
numbers(4.5, 2.1, 7.8)
```

```python
#Define a function that takes an integer number as an input parameter and checks wether the
#number is a prime number
#include a check whether the value is indeed an integer number. If incorrect output a
#message that tells the user what went wrong
#print the result using a formatted print command

import math

def param(integer):
  if integer == int(integer):
    if integer<2:
      print("It is not a prime number")
    elif integer==2:
      print("It is a prime number")
    elif integer%2==0:
      print("It is not a prime number")
    else:
        # Checking odd divisors up to the square root of the number
        limit = int(math.sqrt(integer)) + 1

        is_prime = True
        for i in range(3, limit, 2):
            if integer % i == 0:
                is_prime = False
                break

        if is_prime:
            print(f"The number {integer} is a prime number")
        else:
            print(f"The number {integer} is not a prime number")

  else:
    print("Error: I need a whole number (an integer) to check for primality")



param(2)
param(17)
param(15)
param(4.5)
param(-5)
param(24)
```

```python
#Define a function that takes a string as an input parameter and checks wether the input string is a palindrom (a word or phrase that reads the same backward as it reads forward)
#include a check whether the value is indeed a string. If incorrect output a message that tells the user what went wrong
#print the result using a formatted print command

def palindrome(word):
  if word == str(word):
    if word == word[::-1]:
      print(f"The input {word} is a palindrome")
    else:
        print("The input is not a palindrome")
  else:
    print("The input is not a string")

palindrome(3)
palindrome("room")
palindrome("level")
```

```python
#Define a function that takes an integer number as an input parameter and checks wether the number is a prime number
# include a check whether the value is indeed an integer number. If incorrect output a message that tells the user what went wrong
#print the result using a formatted print command

def is_prime(n):

    if n <= 1:
        return False

    i = 2
    while i * i <= n:
        if n % i == 0:
            return False
        i += 1
    return True

count = 0  # No of primes found
number = 1 # Current integer
prime_100th = 0

# Looping until the 100th prime number is found
while count < 100:
    number += 1

    # Checking if the current number is prime
    if is_prime(number):
        count += 1
        prime_100th = number

print(f"The 100th prime number is: {prime_100th}")
```

```python
#According to legend, the inventor of the chess game and board (8x8 squares) asked his ruler to pay him in rice grains, putting 1 grain on the first square and doubling the number of grains from one square field to the next (i.e. 1 on the first, 2 on the second, 4 on the third, 8 on the fourth square).
#Create a 2D, 8x8 array for the chess board squares. Use a nested for-loop to fill them with grains accordingly.
#How many rice grains will you end up with on the last square?
#Hint: Start with a smaller board, i.e. 3x3, and then go from there to a full chessboard

import numpy as np

Board_size = 8
Total_squares = Board_size * Board_size



chessboard = np.zeros((Board_size, Board_size), dtype=object) #data type is object because of very large numbers

current_grains = 1

total_grains = 0 # Lets keep it as zero for now

# Using for loops
for row in range(Board_size):
    for col in range(Board_size):

        chessboard[row, col] = current_grains
        total_grains += current_grains
        current_grains *= 2

last_square_grains = chessboard[Board_size - 1, Board_size - 1]

print(" Chessboard Grain Distribution,Here I am only showing the first 10 rows due to the size issue ")
print(chessboard[:10])
print(f"The total number of squares processed was: {Total_squares}")
print(f"The number of rice grains on the last square is: {last_square_grains}")
```
